```python
def year(year):
    year_list = []
    for album in data:
        if str(year) == album['year']:
            year_list.append(album['album'])
    return year_list
```

```python
def year_range(year1, year2):
    yrange_list = []
    for album in data:
        if album['year'] >= str(year1) and album['year'] <= str(year2):
            yrange_list.append(album['album'])
    return yrange_list
```
